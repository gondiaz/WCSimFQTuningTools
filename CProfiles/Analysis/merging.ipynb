```python
import os
import re
import uproot
import numpy as np

import matplotlib.pyplot as plt
%matplotlib widget
from matplotlib.colors import LogNorm
```

### Merging

```python
fname  = os.path.expandvars("$LUSTRE/CProfiles/mu-/tuning/unnormalized/cprofiles_merged.root")
f  = uproot.open(fname)
print(f.classnames())

# get momentum values
momenta = np.sort([float(re.findall(r'\d+(?:\.\d+)?', key)[0]) for key in f.keys() if re.match("g_\d+", key)])
```

```python
# choose random momentum
momentum = np.random.choice(momenta)
H, thbins, sbins  =  f[f"g_{momentum}"].to_numpy()

plt.figure(figsize=[5, 3])

plt.title(f"Momentum = {momentum}")
plt.pcolormesh(thbins, sbins, H.T, cmap="jet")
plt.xlim([0.6, 0.9])
plt.ylim([None, 250])
plt.xlabel(r"$\cos \theta$")
plt.ylabel("s [cm]")
plt.colorbar()

plt.tight_layout()
```
