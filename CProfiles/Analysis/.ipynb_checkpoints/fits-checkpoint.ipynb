```python
import re
import uproot
import ROOT
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

def get_tgrapherrors_values(g):
    n = g.GetN()
    x, y = [], []
    for i in range(n): 
        x.append(g.GetPointX(i))
        y.append(g.GetPointY(i))
    return x, y
```

```python
# o stands for "old"
ofname = "../TuningFile/CProf_11_fit_out.root"
fname  = "../TuningFile/cprofiles_fits.root"

of = uproot.open(ofname)
f  = uproot.open(fname)
```

```python
for key in of.keys():
    if not re.match("I_\d+_\d+_\d+", key): print(key)
```

```python
n = 0
# get fit parameters
ojs, or0bins, oth0bins, ojbins = of[f"hI3d_par_{n}"].to_numpy()
js ,  r0bins,  th0bins,  jbins =  f[f"Js_{n}"]      .to_numpy()

np.testing.assert_allclose(or0bins, r0bins)
```

```python
r0bin  = 0
th0bin = 1

# get integrals
fin = ROOT.TFile(ofname)
oe, oI = get_tgrapherrors_values(fin.Get(f"I_{n}_{r0bin+1}_{th0bin+1}"))
Is, r0bins, th0bins, ebins = f[f"I_{n}"].to_numpy()
e = np.log((ebins[1:] + ebins[:-1])/2.)
I = Is[r0bin, th0bin]
```

```python
plt.figure()
plt.scatter(oe, oI, color="red"  , label="old")
plt.scatter( e,  I, color="black", label="new")

es = np.linspace(e[0], e[-1], 100)
p = np.poly1d(ojs[r0bin, th0bin, :6])
plt.plot(es, p(es), color="red"  , label="old")

plt.tight_layout()
```
